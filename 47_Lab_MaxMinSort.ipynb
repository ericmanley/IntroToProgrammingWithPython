## Working with the built-in `sort` method

If we didn't want to write a custom function for student records, we could attempt to use the built-in `sort` method. However, by default, it will sort based on the first item in the tuple, which is the student's name (lexigraphically, using the UNICODE value of the characters in the string):

```python
student_records.sort()
print(student_records)
```

If we instead wanted to sort by the GPA, we'd need to write a special function to extract the GPA from the record, like this:

```python
def get_gpa(student_record):
    return student_record[3]

print("Here's a student's GPA:", get_gpa(("Stu Dent",100123456,"Computer Science",3.4)) )
```

Then, we can pass the *name* of this function `get_gpa` as the optional `key` parameter when we call the `sort` method. Note that it now sorts by GPA.

```python
student_records.sort(key=get_gpa)
print( student_records )
```

And in our case, we wanted it to go from largest to smallest. There's another optional parameter, `reverse`, that we can use for that.

```python
student_records.sort(key=get_gpa, reverse=True)
print( student_records )
```
